```python
import html
import re
import httpx
```

```python
def get_summary(q):
    """Fetches a short summary from Wikipedia based on the search term."""
    response = httpx.get("https://en.wikipedia.org/w/api.php", headers={"manishporwal1987@gmail.com": "ReactAgent/1.0"}, params={
        "action": "query",
        "list": "search",
        "srsearch": q,
        "format": "json"
    })

    raw_snippet = response.json()["query"]["search"][0]["snippet"]

    # Remove HTML tags
    clean_snippet = re.sub(r'<[^>]+>', '', raw_snippet)

    # Unescape HTML entities (e.g., &quot; → ", &amp; → &)
    text_snippet = html.unescape(clean_snippet)

    return text_snippet

summary= get_summary("elon musk")
print(summary)   
```
